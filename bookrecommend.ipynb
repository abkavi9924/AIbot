```python
def strtolist(word):
    L = []
    for i in word:
        if i == '[' or i == ']' or i == "'":
            L = L
        else:
            L.append(i)
    return L
```

```python
bdata['authors']=bdata['authors'].apply(strtolist)
```

```python
bdata['authors']=bdata['authors'].apply(lambda x:"".join(x))
```

```python
bdata['authors']=bdata['authors'].apply(lambda x:x.split(","))
```

```python
bdata['categories']=bdata['categories'].apply(lambda x:"".join(x))
bdata['categories']=bdata['categories'].apply(lambda x:x.split(','))
```

```python
bdata['tags'] = bdata['description'] + bdata['categories'] + bdata['authors'] + bdata['publisher']
```

```python
bdata = bdata[['Title','publishedYear','tags']]
```

```python
bdata.head()
```

```python
cv = CountVectorizer(max_features = 10, stop_words='english')
```

```python
stemmer = stem.PorterStemmer()
```

```python
def stemming(word):
    y = []
    for i in word.split():
        y.append(stemmer.stem(i))
    return " ".join(y)
```

```python
bdata['tags']=bdata['tags'].apply(lambda x:" ".join(x))
```

```python
bdata.head()
```

```python
bdata['tags']=bdata['tags'].apply(lambda x:x.lower())
```

```python
bdata.head()
```

```python
bdata['tags']=bdata['tags'].apply(stemming)
```

```python
bdata.head()
```

```python
va = cv.fit_transform(bdata['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```
